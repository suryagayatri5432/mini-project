# slice a data set

```python
data=bodyfat.drop(['Neck','Chest','Age','Density','Height','Abdomen','Hip','Thigh','Knee','Ankle','Biceps','Forearm','Wrist'],axis=1)
```

```python
data
```

```python
x=data.iloc[:,0].values.reshape(-1,1)
y=data.iloc[:,-1].values.reshape(-1,1)
```

```python
x
```

```python
y
```

```python
plt.scatter(x,y)
plt.xlabel('BodyFat')
plt.ylabel('Weight')
plt.show()
```

# Data Spliting

```python
from sklearn.model_selection import train_test_split
```

```python
x_train,x_test,y_train,y_test= train_test_split(x,y,test_size=0.2,random_state=0)
```

```python
x_test.shape
```

```python
x_train.shape
```

```python
y_train.shape
```

```python
y_test.shape
```

# import syntax

```python
from sklearn.linear_model import LinearRegression
```

# applying linearregression

```python
lm=LinearRegression()
```

```python
lm
```

```python
lm.fit(x_train,y_train)
```

# predict

```python
y_predit=lm.predict(x_test)
```

```python
y_predit
```
